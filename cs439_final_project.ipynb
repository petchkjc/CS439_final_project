```python
from django.contrib.sessions.models import Session
from django.contrib.sessions.backends.db import SessionStore
import json
data=[]
for i in Session.objects.all():
#     i.delete()
    d =SessionStore().decode(i.session_data)
    if d.get('person_id'):
        data.append({
                "person_id": d.get('person_id'),
                "image_id": d.get('image_id')
            })
        print "%s: %s"%(d.get('person_id'), d.get('image_id'))
```

```python
s=[]
for i in data:
    s.append(i.get('image_id'))
print s
```

```python
k=0
while k < len(s):
    j=0
    while j < len(s[k]):
        s[k][j] = int(s[k][j])
        j+=1
        
    k+=1
        
print s
```

```python
A=[]
N=3
for i in s:
    #print i
    M=len(i)
    if M >= N:
        for j in xrange(0,M-N+1):
            #print i[j:j+N]
            A.append(i[j:j+N])
```

```python
import numpy as np
Xy=np.array(A)
Xy.shape
```

```python
pC = []
count = 0
while( count < len( p[0] )):
    pC.append( ( p[0][count] , c[count] ) )
    count += 1


pC.sort(reverse=True,key=lambda tup: tup[0])

numberOfPic = 10


if(len(pC) >= numberOfPic):
    pC = pC[0:numberOfPic]

print pC

predict_id=[]
for i in pC:
    predict_id.append(i[1])
    
print predict_id
```

```python
import unicodedata
new_gender=[]
new_gender2=[]
for t in person_gender:
    new_gender.append(unicodedata.normalize('NFKD', t).encode('ascii','ignore'))
    
code1={'Female':0,'Male':1}
for sb in new_gender:
    sa=code1[sb]
    new_gender2.append(sa)
print new_gender2
```
